# Model Training

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
import joblib
```

```python
df_train = df.copy()
```

```python
x_win_target = df["x_wins"]
is_draw_target = df["is_draw"]
```

```python
X = df_train.drop(columns=["x_wins", "is_draw"])
print(X.shape)
X
```

```python
print(x_win_target.shape)
print(is_draw_target.shape)
```

```python
X_train_win, X_test_win, y_train_win, y_test_win = train_test_split(X, x_win_target, test_size=0.2, random_state=101)
X_train_is_draw, X_test_is_draw, y_train_is_draw, y_test_is_draw = train_test_split(X, is_draw_target, test_size=0.2, random_state=101)
```

```python
ModelsWins = LogisticRegression(max_iter=1000, class_weight='balanced')
ModelsIsDraw = LogisticRegression(max_iter=1000, class_weight='balanced')
```

```python

parametres = {
    'C': [0.001, 0.01, 0.1, 1, 10, 100],
    'penalty': ['l1', 'l2'],
    'solver': ['liblinear', 'saga'] 
}

grid_search_wins = GridSearchCV(
    estimator=ModelsWins,
    param_grid=parametres,
    cv=5,
    scoring=['accuracy', 'f1', 'recall', 'precision'],
    refit='f1',
    n_jobs=-1 
)

grid_search_isDraw = GridSearchCV(
    estimator=ModelsIsDraw,
    param_grid=parametres,
    cv=5,
    scoring=['accuracy', 'f1', 'recall', 'precision'],
    refit='f1',
    n_jobs=-1 
)
```

```python
grid_search_wins.fit(X_train_win, y_train_win)
grid_search_isDraw.fit(X_train_is_draw, y_train_is_draw)
```

```python
meilleur_modele_wins = grid_search_wins.best_estimator_
prediction_win = meilleur_modele_wins.predict(X_test_win)
meilleur_modele_isDraw = grid_search_isDraw.best_estimator_
prediction_isDraw = meilleur_modele_isDraw.predict(X_test_is_draw)

print(f"Meilleurs paramètres pour wins: {grid_search_wins.best_params_}")
print(f"Meilleurs paramètres pour isDraw: {grid_search_isDraw.best_params_}")
print(f"Meilleur score de validation sur win: {grid_search_wins.best_score_:.4f}")
print(f"Meilleur score de validation sur isDraw: {grid_search_isDraw.best_score_:.4f}")

score_test_wins = meilleur_modele_wins.score(X_test_win, y_test_win)
score_test_isDraw = meilleur_modele_isDraw.score(X_test_is_draw, y_test_is_draw)
print(f"Score sur les données de test Wins : {score_test_wins:.4f}")
print(f"Score sur les données de test IsDraw : {score_test_isDraw:.4f}")
```

```python
# Evaluation avec Accuracy
acc_wins = accuracy_score(y_test_win, prediction_win)
acc_isDraw = accuracy_score(y_test_win, prediction_isDraw)

print(f"Précision globale sur x_Wins : {acc_wins:.4f}")
print(f"Précision globale sur is_Draw : {acc_isDraw:.4f}")
```

```python
# Evaluation en utilisant F1-Score
print(classification_report(y_test_win, prediction_win))
print("-----------------------\n")
print(classification_report(y_test_is_draw, prediction_isDraw))
```

```python
# Evaluation en utilisant la confusion metrics
print('Confusion Matrix de x_wins : \n', confusion_matrix(y_test_win, prediction_win))
tn_x, fp_x, fn_x, tp_x = confusion_matrix(y_test_win, prediction_win).ravel()

print('-------------------------------------------')

print('Confusion Matrix de is_Draw : \n', confusion_matrix(y_test_is_draw, prediction_isDraw))
tn_0, fp_0, fn_0, tp_0 = confusion_matrix(y_test_is_draw, prediction_win).ravel()
```
